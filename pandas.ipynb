```python
# drop rows where ANY column has null
df.dropna()

# drop rows where SPECIFIC column has null
df.dropna(subset=['rating'])

# drop rows where ALL columns are null
df.dropna(how='all')

# drop columns that have nulls
df.dropna(axis=1)
```

```python
# ✅ clean chaining approach
(
    df
    .dropna(subset=['director'])
    [['title', 'director', 'country']]
    .head()
)
```

```python
# fill with a static value
df['rating'].fillna('Not Rated')

# text columns → fill with 'Unknown' or most frequent value
df['country'].fillna('Unknown')
df['rating'].fillna(df['rating'].mode()[0])  # most frequent rating

# fill with mean/median — for numeric columns
df['release_year'].fillna(df['release_year'].mean())

# numeric columns → fill with median (more robust than mean)
df['release_year'].fillna(df['release_year'].median())

# fill with forward fill — use previous row's value
df['date_added'].fillna(method='ffill')

# fill with backward fill — use next row's value
df['date_added'].fillna(method='bfill')
```

```python
# rename specific columns
# 💼 Industry Practice — rename while chaining
(
    df.rename(columns={'listed_in': 'genre'
                       'date_added': 'added_date'
                       })
    [['title', 'genre', 'type']]
    .head()
)
```

```python
# make all columns lowercase, replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# drop single column
df.drop(columns=['show_id'])

# drop multiple columns
df.drop(columns=['show_id', 'description'])

# drop specific rows by index
df.drop(index=[0, 1, 2])

# 💼 Industry Practice — chain it
(
    df.drop(columns=['show_id', 'description'])
    .head()
)
```

```python
# check current dtypes
df.dtypes

# convert release_year from int to string
df['release_year'].astype(str)

# convert a column to category — saves memory on repetitive text
df['type'].astype('category')
df['rating'].astype('category')
```

```python
# check duplicates
df.duplicated().sum()

# check duplicates on specific column
df.duplicated(subset=['title']).sum()

# view duplicate rows
df[df.duplicated(subset=['title'])]

# drop duplicates
df.drop_duplicates()

# drop duplicates on specific column, keep last occurrence
df.drop_duplicates(subset=['title'], keep='last')
```

```python
# convert date_added from string to datetime
df['date_added'] = pd.to_datetime(df['date_added'])

# now you can extract parts
df['date_added'].dt.year
df['date_added'].dt.month
df['date_added'].dt.day_name()   # Monday, Tuesday...

# 💼 Industry Practice — create new columns from dates
(
    df.assign(
        year_added = df['date_added'].dt.year,
        month_added = df['date_added'].dt.month
    )
    [['title', 'date_added', 'year_added', 'month_added']]
    .head()
)
```

```python
# remove leading/trailing spaces
df['title'].str.strip()

# convert to lowercase
df['title'].str.lower()

# replace within string
df['duration'].str.replace(' min', '').str.replace(' Season', 'S')

# split a column — duration has '90 min' or '2 Seasons'
df['duration'].str.split(' ', expand=True)
#    0       1
# 0  90     min
# 1  2   Seasons
```
