```python
# 💼 Industry Practice — a full clean pipeline
(
    df
    # 1. fix dtypes first
    .assign(date_added = pd.to_datetime(df['date_added']))
    .astype({'type': 'category', 'rating': 'category'})

    # 2. clean nulls
    .dropna(subset=['country', 'rating'])
    .fillna({'director': 'Unknown', 'cast': 'Unknown'})

    # 3. transform
    .assign(
        era          = pd.cut(df['release_year'],
                              bins=[1900,1999,2009,2021],
                              labels=['Classic','2000s','2010s']),
        year_added   = lambda x: x['date_added'].dt.year,
        content_type = np.where(df['type'] == 'Movie', 'Film', 'Series')
    )

    # 4. filter
    .query("release_year >= 2000")

    # 5. select only needed columns
    [['title', 'content_type', 'era', 'year_added', 'country', 'rating']]
)
```

```python
# define steps as functions
def fix_dtypes(df):
    return df.assign(date_added = pd.to_datetime(df['date_added']))

def clean_nulls(df):
    return df.dropna(subset=['country', 'rating'])

def add_features(df):
    return df.assign(
        year_added   = df['date_added'].dt.year,
        content_type = np.where(df['type'] == 'Movie', 'Film', 'Series')
    )

# connect with .pipe() — reads like a story
(
    df
    .pipe(fix_dtypes)
    .pipe(clean_nulls)
    .pipe(add_features)
    .head()
)
```

```python
# 1. LOAD
df = pd.read_csv('netflix_titles.csv', parse_dates=['date_added'])

# 2. INSPECT
df.shape
df.info()
df.isnull().sum()
df.describe()

# 3. CLEAN
df = (
    df
    .drop_duplicates()
    .dropna(subset=['country', 'rating'])
    .fillna({'director': 'Unknown', 'cast': 'Unknown'})
    .astype({'type': 'category', 'rating': 'category'})
)

# 4. TRANSFORM
df = (
    df
    .assign(
        year_added   = df['date_added'].dt.year,
        content_type = np.where(df['type'] == 'Movie', 'Film', 'Series'),
        era          = pd.cut(df['release_year'],
                              bins=[1900,1999,2009,2021],
                              labels=['Classic','2000s','2010s'])
    )
)

# 5. ANALYSE
(
    df.groupby(['content_type', 'era'])
    .agg(total = ('title', 'count'))
    .reset_index()
    .sort_values('total', ascending=False)
)
```
